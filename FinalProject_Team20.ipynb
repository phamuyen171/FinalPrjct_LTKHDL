##### *3.5 Kiểu dữ liệu của từng cột là gì? Có cột nào có kiểu dữ liệu không phù hợp không?* <a id="3-5"></a>

```python
print (raw_df.dtypes)
```

Ta nhận thấy, cột `reviewcount` nên có kiểu dữ liệu là kiểu số thay vì object, nên ta sẽ tiến hành thay đổi nó. Nhưng trước tiên, vì giá trị số của cột đang tồn tại ở dạng đặc biệt (ví dụ: 100k) nên ta cần chuẩn hóa nó trước khi chuyển đổi kiểu dữ liệu

```python
def convert_to_number(value):
    if isinstance(value, str): 
        value = value.strip() 
        if 'k' in value.lower():  
            return float(value.lower().replace('k', '')) * 1000
        else:
            return float(value)  
    return value 

raw_df['reviewcount'] = raw_df['reviewcount'].apply(convert_to_number)
raw_df['reviewcount'] = pd.to_numeric(raw_df['reviewcount'], errors='coerce')

print("\nKiểu dữ liệu sau khi chuẩn hóa:")
print(raw_df['reviewcount'].dtypes)
print (raw_df['reviewcount'])
```

##### *3.6. Đối với các cột số, tỷ lệ phần trăm giá trị thiếu bằng bao nhiêu? Giá trị lớn nhất và nhỏ nhất là?* <a id="3-6"></a>

```python
def analyze_numerical_columns(df):
    def find_outliers(series):
        q1 = series.quantile(0.25)  
        q3 = series.quantile(0.75)  
        iqr = q3 - q1               
        lower_bound = q1 - 1.5 * iqr  
        upper_bound = q3 + 1.5 * iqr  

        outliers = series[(series < lower_bound) | (series > upper_bound)]
        return outliers

    numerical_columns = df.select_dtypes(include=['number']).columns 
    analysis_results = []
    outliers_dict = {}  

    for col in numerical_columns:
        col_data = df[col]
        missing_percentage = col_data.isna().mean() * 100
        col_min = col_data.min()
        col_max = col_data.max()
        description = col_data.describe()

        outliers = find_outliers(col_data)
        outliers_dict[col] = outliers

        analysis_results.append({
            'Column': col,
            'Missing (%)': missing_percentage,
            'Min': col_min,
            'Max': col_max,
            'Outliers Count': len(outliers),
            'Outliers Percentage (%)': len(outliers) / len(col_data.dropna()) * 100,
        })

    return pd.DataFrame(analysis_results), outliers_dict
```

```python
results, outliers = analyze_numerical_columns(raw_df)
print("Phân tích cột có dạng số: ")
print(results)

print("\nGiá trị bất thường:")
for col, outlier_values in outliers.items():
    print(f"Cột '{col}':")
    print(outlier_values)
```

Vậy cột có dạng số được phân tích ở đây là `rating` và `reviewcount`

- Với cột `rating`, tỷ lệ thiếu giá trị đạt 12.04%, và đạt 24 giá trị outliers
- Với cột `reviewcount`, tỷ lệ thiếu giá trị đạt 12.04%, và đạt 141 giá trị outliers

##### *3.7 Đối với các cột category, tỷ lệ phần trăm giá trị thiếu bằng bao nhiêu?* <a id="3-7"></a>

```python
def missing_percentage_categorical(df):
    
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    
    missing_percentages = {}
    for col in categorical_columns:
        missing_percentages[col] = df[col].isna().mean() * 100
        
    result_df = pd.DataFrame({
        "Column": list(missing_percentages.keys()),
        "Missing (%)": list(missing_percentages.values())
    })
    
    return result_df
```

```python
print (missing_percentage_categorical(raw_df))
```

Như vậy, cột `partner` và cột `course` không có giá trị thiếu, cột đạt tỷ lệ giá trị thiếu cao nhất là cột `reviewcount`

**Mỗi cột như vậy có bao nhiêu giá trị phân biệt, nêu một vài giá trị mẫu?** 

```python
def analyze_categorical_columns(df):

    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    results = []
    
    for col in categorical_columns:
        unique_count = df[col].nunique()               
        sample_values = df[col].dropna().unique()[:]
        
        results.append({
            'Column': col,
            'Unique Values Count': unique_count,
            'Sample Values': sample_values
        })
    
    result_df = pd.DataFrame(results)
    return result_df
```

```python
print (analyze_categorical_columns(raw_df))
```

**Liệu có giá trị nào bất thường hay không?**

```python
def abnormal_values_all_columns(df):
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    abnormal_values_dict = {}
    
    for col in categorical_columns:
        abnormal_values = df[df[col].isna() | (df[col].str.strip() == '')][col]
        
        if not abnormal_values.empty:
            abnormal_values_dict[col] = abnormal_values.unique()
    
    return abnormal_values_dict
```

```python
abnormal = abnormal_values_all_columns(raw_df)
for column, values in abnormal.items():
    print(f"- {column}: {values}")
```

Như vậy, tất cả các giá trị bất thường trong các cột category đều là các giá trị bị thiếu.

```python
# điền vào các cell trống giá trị chuỗi rỗng
if df['skills'].isna().sum() != 0:
    df['skills'] = df['skills'].fillna('')
# chuyển kiểu chuỗi của cột 'skills' thành list
if not all(df['skills'].apply(lambda x: isinstance(x, list))):
    df['skills'] = [skill_list
                    .replace('{', '')
                    .replace('}', '')
                    .replace('"', '')
                    .split(',') for skill_list in df['skills']]

df['skills']
```

**Phân loại các khóa học theo từng nhóm ngành cụ thể.**

```python
# tạo từ điển phân loại lưu các keywords của chủ đề cụ thể
categories_dict = {
    'Technology' :[
        'Cloud', 'Python', 'Programming', 'Machine Learning', 'Data', 'SQL', 
        'Linux', 'Computer', 'AI', 'Cyber', 'Security', 'Network', 'Software', 'DevOps'
    ],
    'Science': [
        'Statistics', 'Biostatistics', 'Mathematics', 'Physics', 'Chemistry', 
        'Research', 'Epidemiology', 'Neural', 'Bioinformatics'
    ],
    'Finance': [
        'Accounting', 'Budget', 'Investment', 'Financial', 'Tax', 
        'Audit', 'Procurement', 'Securities', 'Cash', 'Management'
    ],
    'Business': [
        'Leadership', 'Marketing', 'Sales', 'Management', 'HR', 
        'Business', 'Entrepreneurship', 'Customer', 'Operations', 'Planning'
    ],
    'Health': [
        'Health', 'Nutrition', 'Public Health', 'Clinical', 'Medical'
    ],
    'Society': [
        'Social', 'Human', 'Culture', 'Psychology'
    ],
}
```

```python
# hàm phân loại skill
# xây dựng hàm phân loại dựa trên các keywords càng nhiều các 
# keywords thuộc về một chủ đề thì chủ đề đó là chủ đề chính của khóa học
def classify_skill(skill_list, cate = categories_dict):
    if len(skill_list) == 1 and skill_list[0] == '':
        return None
    cnt_cate = {cate_key: 0 for cate_key in cate.keys()}

    category_keywords = {
        word: category for category, keywords in cate.items() for word in keywords
        }

    for skill in skill_list:
        for word in skill.split():
            if word in category_keywords:
                cnt_cate[category_keywords[word]] += 1

    sorted_cnt = dict(sorted(cnt_cate.items(), key=lambda item: item[1], reverse=True))
    return list(sorted_cnt.keys())[0]
```

```python
# thêm cột phân loại vào df 
df['Category'] = df['skills'].apply(lambda skills: classify_skill(skills, categories_dict))
```

```python
duration_mapping = {
    "months": "Months",
    "meses": "Months",
    "year": "Year",
    "years": "Years",
    "hour": "Hour",
    "hours": "Hours",
    "to":"-",
    "or":"-",
    "- complete":"",
    "\.": "",
    "–": "-",
}


for key, value in duration_mapping.items():
    df_copy["duration"] = df_copy["duration"].str.replace(key, value, regex=True, case=False)

# Thêm khoảng trắng trước và sau dấu "-"
df_copy["duration"] = df_copy["duration"].str.replace(r"(\d+)-(\d+)", r"\1 - \2", regex=True)

# Thay thế "a" thành "-" trong "18 a 24 Months"
df_copy["duration"] = df_copy["duration"].str.replace(r"(\d+)\s*a\s*(\d+)\s*Months", r"\1 - \2 Months", regex=True)

# Loại bỏ phần thừa trong chuỗi "2 - 4 Years depending on ..."
df_copy["duration"] = df_copy["duration"].str.replace(r"(\d+ - \d+ Years).*", r"\1", regex=True)

df_copy["duration"] = df_copy["duration"].str.strip()
```

Bởi vì chưa chắc khóa học nhận về ít lượt đánh giá thì điểm đánh giá của nó sẽ thấp, hay khóa học nhận về nhiều lượt đánh giá thì chất lượng khóa học sẽ cao.
- Thống tin này không những hữu ích cho người dùng mà còn hữu ích cho những đối tác cung cấp khóa học trong việc cải thiện chất lượng.


**Trước hết, ta cần kiểm tra xem các cột `rating` và `reviewcount` có ở định dạng số thực chưa?**

```python
print (df['rating'].dtypes)
print (df['reviewcount'].dtypes)
```

**Sau đó ta cần loại bỏ các giá trị NaN trong hai cột này**

```python
df_clean = df.dropna(subset=['rating', 'reviewcount'])
```

**Ta xét  top 10 khóa học đạt rating cao nhất và top 10 khóa học có nhiều lượt review nhất**

```python
top_10_rating = df_clean.sort_values(by='rating', ascending=False).head(10)
top_10_reviewcount = df_clean.sort_values(by='reviewcount', ascending=False).head(10)

fig, axes = plt.subplots(1, 2, figsize=(18, 6)) 

# Biểu đồ 1: Top 10 khóa học có rating cao nhất
sns.barplot(x='rating', y='course', data=top_10_rating, palette='viridis', ax=axes[0])
axes[0].set_title('Top 10 Khóa học có Rating cao nhất', fontsize=14)
axes[0].set_xlabel('Rating', fontsize=12)
axes[0].set_ylabel('Tên khóa học', fontsize=12)

# Biểu đồ 2: Top 10 khóa học có nhiều lượt đánh giá nhất
sns.barplot(x='reviewcount', y='course', data=top_10_reviewcount, palette='magma', ax=axes[1])
axes[1].set_title('Top 10 Khóa học có nhiều Lượt đánh giá nhất', fontsize=14)
axes[1].set_xlabel('Số lượt đánh giá', fontsize=12)
axes[1].set_ylabel('Tên khóa học', fontsize=12)

plt.tight_layout() 
plt.show()
```

Ta nhận thấy top10 khóa học có điểm đánh giá cao và top10 khóa học nhận được nhiều lượt đánh giá nhất hoàn toàn khác biệt. Điều này phần nào đó đã khẳng định sự ÍT ảnh hưởng lẫn nhau giữa hai thuộc tính `rating` và `reviewcount`. Để có được sự đánh giá dựa trên toàn bộ tập dữ liệu, ta cần dựa theo sự phân tích dựa trên việc tính hệ số tương quan của hai thuộc tính này.

**Tính hệ số tương quan giữa hai thuộc tính**

```python
correlation = df_clean['rating'].corr(df_clean['reviewcount'])
print(f"Hệ số tương quan giữa rating và reviewcount: {correlation:.2f}")
```

Hệ số tương quan xấp xỉ 0 cho thấy gần như không có mối liên hệ tuyến tính giữa số lượt đánh giá khóa học và điểm đánh giá khóa học đó.

**Trực quan hóa kết quả phân tích trên**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='reviewcount', y='rating', data=df_clean, alpha=0.7)
plt.title('Mối quan hệ giữa xếp hạng khóa học và số lượng đánh giá')
plt.xlabel('Số lượng đánh giá')
plt.ylabel('Xếp hạng khóa học')
plt.grid(True)
plt.show()
```

##### **Q5:** *Đối tác nào cung cấp đa dạng các khóa học cho các trình độ của người học nhất?* <a id="q5"></a>

**Ý nghĩa câu hỏi** Đánh giá độ đa dạng của các đối tác trong việc cung cấp khóa học ở nhiều trình độ khác nhau, cũng nhìn ra được đối tác nào chuyên cung cấp khóa học cho trình độ người học nào

**Tiền xử lý dữ liệu hai cột `partner` và `level`**

```python
df_clean = df.dropna(subset=['partner', 'level'])
df_clean['level'] = df_clean['level'].str.strip()
```

**Các cấp bậc trình độ người học có trong dataset**

```python
print (df_clean['level'].unique())
```

**Tính số lượng trình độ khác nhau mà mỗi đối tác cung cấp**

```python
# Tính số lượng trình độ khác nhau cho mỗi đối tác
diversity_by_partner = df_clean.groupby('partner')['level'].nunique().reset_index()
diversity_by_partner.columns = ['partner', 'unique_levels']
diversity_by_partner = diversity_by_partner.sort_values(by='unique_levels', ascending=False)

print("Số lượng trình độ khác nhau cho mỗi đối tác:")
print(diversity_by_partner)
```

**Tính số lượng khóa học của từng cấp bậc trình độ người học mà từng đối tác cung cấp**

```python
partner_level_counts = df_clean.groupby(['partner', 'level']).size().reset_index(name='count')
partner_level_counts
```

**Xem xét sự đa dạng về trình độ người học của top10 đối tác cung cấp nhiều khóa học nhất**

```python
top_10_partners = df_clean['partner'].value_counts().head(10).index
df_top_10 = df_clean[df_clean['partner'].isin(top_10_partners)]

partner_level_counts = df_top_10.groupby(['partner', 'level']).size().reset_index(name='count')

pivot_data = partner_level_counts.pivot(index='partner', columns='level', values='count').fillna(0)
pivot_data['total_courses'] = pivot_data.sum(axis=1)  
pivot_data = pivot_data.sort_values(by='total_courses', ascending=False) 
pivot_data = pivot_data.drop(columns='total_courses')

pivot_data.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(12, 8))

plt.xticks(rotation=45, ha='right')
plt.title('Top 10 đối tác có sự đa dạng về trình độ khóa học (sắp xếp giảm dần)')
plt.xlabel('Đối tác (partner)')
plt.ylabel('Số lượng khóa học (number of courses)')
plt.legend(title='Trình độ khóa học', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Ta nhận thấy, Google là đối tác cung cấp nhiều khóa học nhất nhưng chỉ cung cấp cho hai cấp bậc là Advanced và Beginner. Hầu hết các đối tác còn lại ngoại trừ Google đều cung cấp khóa học cho từ 3 trình độ trở lên.

##### **Q6:** *Các khóa học có chứng chỉ (Professional Certificate) khác biệt gì so với các khóa học khác về xếp hạng và thời lượng?* <a id="q6"></a>

**Ý nghĩa câu hỏi:** Giúp hiểu rõ giá trị của chứng chỉ trong các khóa học và liệu nó có ảnh hưởng đến chất lượng (xếp hạng) và thời lượng khóa học hay không.

**Tiền xử lý dữ liệu**

```python
df_clean = df.dropna(subset=['rating', 'duration', 'certificatetype'])
```

**Trước tiên, chúng ta sẽ nhìn qua số lượng các khóa học có chứng chỉ so với số lượng khóa học không có chứng chỉ**

```python
df_clean.loc[:, 'certificate_group'] = df_clean['certificatetype'].apply(
    lambda x: 'Có chứng chỉ' if 'Professional Certificate' in x else 'Không có chứng chỉ'
)

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='certificate_group', data=df_clean)
for i, patch in enumerate(ax.patches):
    if i == 0: 
        patch.set_facecolor('salmon')
    else: 
        patch.set_facecolor('lightblue') 

plt.title('Số lượng khóa học có chứng chỉ và không có chứng chỉ')
plt.xlabel('Loại khóa học')
plt.ylabel('Số lượng khóa học')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Biểu đồ cột cho thấy, số lượng khóa học có chứng chỉ chỉ bằng 1/5 số lượng khóa học không có chứng chỉ

**So sánh điểm đánh giá giữa khóa học có chứng chỉ và không có chứng chỉ**

```python
df_clean.loc[:, 'certificate_group'] = df_clean['certificatetype'].apply(
    lambda x: 'Có chứng chỉ' if 'Professional Certificate' in x else 'Không có chứng chỉ'
)
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='certificate_group', y='rating', data=df_clean)
plt.title('So sánh xếp hạng giữa khóa học có chứng chỉ và không có chứng chỉ')
plt.xlabel('Loại khóa học')
plt.ylabel('Xếp hạng (Rating)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Biểu đồ boxplot thể hiện rằng đa số các khóa học cho dù là có chứng chỉ hay không có chứng chỉ thì đều nhận được đánh giá cao. Tuy nhiên, các khóa học không có chứng chỉ tồn tại các outliers là các khóa học có đánh giá thấp, và số lượng outliers nhiều hơn rõ rệt so với các khóa học có chứng chỉ

**So sánh thời lượng khóa học giữa khóa học có chứng chỉ và không có chứng chỉ**

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='certificate_group', y='duration', data=df_clean, hue='certificate_group', palette='coolwarm')
plt.title('So sánh thời lượng khóa học giữa khóa học có chứng chỉ và không có chứng chỉ')
plt.xlabel('Loại khóa học')
plt.ylabel('Thời lượng khóa học (giờ)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Rất dễ dàng nhận thấy, các khóa học có chứng chỉ hầu hết đều thời gian nằm trong khoảng từ 3-6 tháng, cho thấy mức độ chuyên nghiệp cũng như đào tạo chuyên sâu trong các khóa học này. Thực thế cho thấy, thời gian này cũng phù hợp với thời gian của một học phần ở các trường Đại học. Các khóa học không có chứng chỉ thì sẽ có các mức thời gian học rộng hơn, người học có thể lựa chọn khóa học phù hợp với mình
